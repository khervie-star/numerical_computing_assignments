```python
import time
import math
import csv
import pandas as pd
```

```python
# Defining our running function f(x) as f(x) = xy
def func(a, b):
    return a * b


x_values = []
y_values = []
y_i = 0
```

```python
###########################################################
#             Inputing our values the hard way            #
###########################################################

def no_csv_function():
    print("======================================================================")
    print("=                       STARTING INTERPOLATION                       =")
    print("======================================================================")
    time.sleep(.4)
    n = eval(input("Enter number of data points 'n': "))

    for i in range(n):
        x_values.append(eval(input("Enter value x(" + str(i+1) + "): ")))

    y_values.append(eval(input("Enter initial value of y 'y1': ")))

    h = eval(input("Enter step-size 'h' of x values: "))
    print("======================================================================")
    print("======================================================================")
    print('')

    for i in range(n):
        s1 = func(x_values[i], y_values[i])
#         print(s1)
        s2 = func(x_values[i] + h/2, y_values[i] + (h/2) * s1)
#         print(s2)
        s3 = func(x_values[i]  + h/2, y_values[i] + (h/2) * s2)
#         print(s3)
        s4 = func(x_values[i] + h, y_values[i] +h * s3)
#         print(s4)

        s = (s1 + (2 * s2) + (2 * s3) + s4) / 6
        print('s: ' + str(s))
        time.sleep(.5)

        y_i = y_values[i] + (h * s)
        y_values.append(y_i)
        print('y('+str(i+1)+'): ' + str(y_i))
        print("======================================================================")
        print("======================================================================")
        print('')  
        time.sleep(.5)
        
    print("Hence for range of values xi to xf, our respcective y-values are: ")
    for i in range(n):
        print(y_values[i+1])
```
